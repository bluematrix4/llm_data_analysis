```python
import pandas as pd

# Load data from csv file
df = pd.read_csv('iplauction2023.csv')

# Now, df is your pandas DataFrame
print(df)
```

```python
print(df.columns.to_list())
```

```python
average_final_price = df.groupby('nationality')['final price (in lacs)'].mean()
print(average_final_price)
```

```python
import matplotlib.pyplot as plt

# Create plot
average_final_price.plot(kind='bar')
plt.ylabel('Average Final Price (in lacs)')  # Set y-axis label
plt.xlabel('Nationality')  # Set x-axis label
plt.title('Average Final Price by Nationality')  # Set plot title
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.show()  # Display the plot
```

```python
df_new = df.groupby(['nationality','player style'])['base price (in lacs)'].mean().reset_index()
df_new = df_new.pivot(index='nationality', columns='player style', values='base price (in lacs)')
print(df_new)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8)) # Set the figure size to make the heatmap more readable
sns.heatmap(df_new, cmap='inferno', annot=True, fmt=".2f")
plt.title('Heatmap of Average Base Price by Nationality and Player Style')  # Set plot title
plt.show()   # Display the heatmap
```
